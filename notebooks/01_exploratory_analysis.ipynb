This notebook provides an exploratory analysis of Alzheimer's disease prevalence and projections
across 8 Latin American countries (2000–2050), using the latamdata-py package.

### Countries covered
Argentina, Bolivia, Brazil, Chile, Colombia, Mexico, Peru, Venezuela

### Topics
1. Install & load data
2. Prevalence trends over time
3. Country comparison
4. Projections to 2050

````python
import latamdata as ld
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

# Plotting style
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.dpi'] = 120
sns.set_theme(style='whitegrid', palette='colorblind')

print(f'latamdata-py version: {ld.__version__}')
````

## 1. Load the Dataset

````python
# Load the full dataset
df = ld.load('alzheimer-latinamerica-dashboard')
print(f'Shape: {df.shape}')
print(f'Countries: {df["country"].unique().tolist()}')
print(f'Year range: {df["year"].min()} – {df["year"].max()}')
df.head(10)
````

````python
# Dataset metadata
info = ld.info('alzheimer-latinamerica-dashboard')
print('Title:', info['title'])
print('DOI:', info['doi'])
print('License:', info['license'])
print('Coverage:', info['coverage'])
````

## 2. Prevalence Trends Over Time

````python
fig, ax = plt.subplots(figsize=(14, 7))

countries = df['country'].unique()
for country in countries:
    country_df = df[df['country'] == country].sort_values('year')
    ax.plot(country_df['year'], country_df['prevalence_rate'],
            marker='o', markersize=3, label=country_df['country_name'].iloc[0])

ax.set_title("Alzheimer's Prevalence Rate in Latin America (2000–2050)", fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Prevalence Rate (% of population 60+)')
ax.legend(loc='upper left', framealpha=0.9)
ax.axvline(x=2024, color='red', linestyle='--', alpha=0.5, label='Current year')
plt.tight_layout()
plt.savefig('prevalence_trends.png', dpi=150, bbox_inches='tight')
plt.show()
````

## 3. Country Comparison (2024)

````python
df_2024 = df[df['year'] == 2024].copy()
df_2024 = df_2024.sort_values('prevalence_rate', ascending=True)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(df_2024['country_name'], df_2024['prevalence_rate'],
               color=sns.color_palette('colorblind', len(df_2024)))

# Add value labels
for bar, val in zip(bars, df_2024['prevalence_rate']):
    ax.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
            f'{val:.1f}%', va='center', ha='left')

ax.set_title("Alzheimer's Prevalence by Country (2024)", fontsize=14, fontweight='bold')
ax.set_xlabel('Prevalence Rate (% of population 60+)')
plt.tight_layout()
plt.savefig('country_comparison_2024.png', dpi=150, bbox_inches='tight')
plt.show()
````
