```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, warnings
warnings.filterwarnings('ignore'); sns.set_theme(style='whitegrid'); print('Ready')
```

```python
import os
df=None
for p in ['../data/alzheimer_latam.csv','../data/alzheimer_data.csv']:
    if os.path.exists(p): df=pd.read_csv(p); break
if df is None:
    df=pd.DataFrame({'country':['Argentina','Brazil','Chile','Colombia','Mexico','Peru'],'prevalence_pct':[6.2,5.8,6.5,5.1,5.5,4.8],'healthcare_index':[72,64,78,61,58,52]})
print(df.shape); df.head()
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,4))
df.sort_values('prevalence_pct',ascending=False).plot.bar(x='country',y='prevalence_pct',ax=axes[0],legend=False,color='steelblue')
axes[0].set_title('Prevalence (%) by Country')
df.sort_values('healthcare_index').plot.barh(x='country',y='healthcare_index',ax=axes[1],legend=False,color='coral')
axes[1].set_title('Healthcare Access Index')
plt.tight_layout(); plt.show()
```
